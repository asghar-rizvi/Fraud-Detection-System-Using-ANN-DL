The data is highly imbalanced which can result in wrong predictions

# First We Balanced The Data

Using Build In Library For Over Sampling Because its uses K nearest algorithm and gives best result

```python
from imblearn.over_sampling import SMOTE
x = df.drop(columns=['Class'])
y = df.Class

smote = SMOTE(sampling_strategy = 'minority')
x_smote, y_smote = smote.fit_resample(x,y)
```

```python
x_smote.shape
```

```python
class_counts = y_smote.value_counts()

plt.figure(figsize=(3, 3))
plt.bar(class_counts.index, class_counts.values, color=['skyblue', 'salmon'])
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Balanced Using Smote')
plt.xticks(class_counts.index)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x['Amount'] = scaler.fit_transform(x[['Amount']])
```

**Spliting data into train and test data sets**

```python
x_train, x_test, y_train, y_test = train_test_split(x_smote, y_smote, train_size=0.8)
```

# Using Machine Learning Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
lrc = LogisticRegression(solver='liblinear', penalty='l1',max_iter=5000)
rfc = RandomForestClassifier(n_estimators=50, random_state=2)
```

```python
from sklearn.metrics import accuracy_score, precision_score
```

```python
dtc.fit(x_train,y_train)
```

```python
accuracy = []
pre = []
```

```python
y_predict = dtc.predict(x_test)
a = accuracy_score(y_test, y_predict)
p = precision_score(y_test,y_predict)
accuracy.append(a)
pre.append(p)
```

```python
lrc.fit(x_train,y_train)
```

```python
y_predict = lrc.predict(x_test)
a = accuracy_score(y_test, y_predict)
p = precision_score(y_test,y_predict)
accuracy.append(a)
pre.append(p)
```
